# Data Preview

Quick inspection notebook for the synthetic gambling dataset.

```python
import pandas as pd
from pathlib import Path

# Resolve path from notebook location so it works regardless of current working directory.
PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
DATA_PATH = PROJECT_ROOT / "data" / "raw" / "synthetic_gambling_data.csv"

df = pd.read_csv(DATA_PATH)

print(f"Loaded: {DATA_PATH}")
print(f"Shape: {df.shape}")
```

```python
pd.set_option("display.max_columns", None)
df.head()
```

```python
columns_df = pd.DataFrame({"column_name": df.columns})
columns_df.index = columns_df.index + 1
columns_df
```

```python
missing_summary = (
    df.isna()
      .sum()
      .sort_values(ascending=False)
      .rename("missing_count")
      .to_frame()
)
missing_summary["missing_pct"] = (missing_summary["missing_count"] / len(df) * 100).round(2)
missing_summary
```
